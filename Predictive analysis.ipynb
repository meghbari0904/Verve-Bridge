```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
train = data.drop(['age','fbs','trestbps'], axis=1)
test = data['age']
```

```python
X_age, X_fbs, y_trestbps, y_chol = train_test_split(train, test, test_size=0.3, random_state=2)
```

```python
regr = LinearRegression()
```

```python
regr.fit(X_age, y_trestbps)
```

```python
pred = regr.predict(X_fbs)
```

```python
pred
```

```python
regr.score(X_fbs, y_chol)
```
